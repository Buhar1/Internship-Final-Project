Weather Prediction Model - Buhari Adamu

A predictive model that uses time features and lagged features to predict temperature and humidity.
The data produced is feed into the Demand Predictive Model.

```python
import pandas as pd
```

```python
file_name1="Demand_Dataset_Post_Prep.csv"

try:
    weather_df=pd.read_csv(file_name1)
    print(f"The file \"{file_name1}\" was loaded successfully.")
except Exception as e:
    print(f"An error occurred while loading the file \"{file_name1}\":\n{e}")
```

```python
weather_df2=weather_df.drop(columns=["Demand", "Demand_Lag_1H", "Demand_Lag_24H", "Demand_Lag_1W", "Demand_Lag_1M", "Rolling_Mean_24H_(Demand)", "Rolling_Std_24H_(Demand)"])
```

```python
weather_df2
```

```python
weather_df2["Timestamp"]=pd.to_datetime(weather_df2["Timestamp"])
```

```python
weather_df2=weather_df2.set_index("Timestamp")
```

```python
weather_df2
```

```python
weather_df2.insert(9, "Temperature_Lagged_1H", weather_df2["Temperature"].shift(1))
```

```python
weather_df2.insert(10, "Temperature_Lagged_1D", weather_df2["Temperature"].shift(24))
```

```python
weather_df2.insert(11, "Temperature_Lagged_1W", weather_df2["Temperature"].shift(168))
```

```python
weather_df2.insert(12, "Temperature_Lagged_1M", weather_df2["Temperature"].shift(672))
```

```python
weather_df2.insert(14, "Humidity_Lagged_1H", weather_df2["Humidity"].shift(1))
```

```python
weather_df2.insert(15, "Humidity_Lagged_1D", weather_df2["Humidity"].shift(24))
```

```python
weather_df2.insert(16, "Humidity_Lagged_1W", weather_df2["Humidity"].shift(168))
```

```python
weather_df2.insert(17, "Humidity_Lagged_1M", weather_df2["Humidity"].shift(672))
```

```python
weather_df2
```

```python
weather_df3=weather_df2.dropna()
```

```python
weather_df3[weather_df3.isna().any(axis=1)]
```

```python
X=weather_df3.drop(columns=["Temperature", "Humidity"])
```

```python
Y1=weather_df3.Temperature

Y1=Y1.to_frame()


Y2=weather_df3.Humidity

Y2=Y2.to_frame()


Y=pd.concat([Y1, Y2], axis=1)
```

```python
Y_train=Y.loc[:"2023-12-01"]

Y_test=Y.loc["2024-01-01":]
```

Model Implementation

```python
import xgboost as xgb

from xgboost import XGBRegressor

from sklearn.model_selection import TimeSeriesSplit

from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
weather_predictive_model=XGBRegressor(
    n_estimators=1000,
    early_stopping_rounds=50,
    learning_rate=0.01,
    random_state=50,
    objective="reg:squarederror"
    
)
```

```python
weather_predictive_model.fit(X_train, Y_train, eval_set=[(X_train, Y_train),(X_test, Y_test)], verbose=False)
```

```python
import numpy as np
```

```python
predictions=weather_predictive_model.predict(X_test)

sqrt_mse=np.sqrt(mean_squared_error(predictions, Y_test))

mae=mean_absolute_error(predictions, Y_test)

print(f"Square Root of Mean Squared Error: {sqrt_mse}\nMean Absolute Error: {mae}")
```

```python
file_name="Weather_Predictive_Model.json"

try:
    weather_predictive_model.save_model(file_name)
    print(f"The model has been successfully saved as \"{file_name}\".")
except Exception as e:
    print(f"An error occurred while saving the model:\n{e}")


model=xgb.XGBRegressor()

model.load_model("Weather_Predictive_Model.json")
```
